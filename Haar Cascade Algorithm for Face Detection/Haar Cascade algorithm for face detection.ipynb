# Haar Cascade algorithm for face detection

```python
import cv2
face=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
image=cv2.imread('MNS.jpg')
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
faces=face.detectMultiScale(gray_image,scaleFactor=1.15,minNeighbors=5)
for x,y,w,h in faces:
    image=cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),5)
print(faces)
print(type(faces))
img_resize=cv2.resize(image,(int(image.shape[1]/2),int(image.shape[0]/2)))
cv2.imshow("Face Detector",img_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
# As an output screenshot is given.
```
